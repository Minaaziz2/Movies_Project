```python
#importing packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading the csv file
df = pd.read_csv('movies_complete.csv', parse_dates=['release_date'])
```

```python
#exploring the dataframe
df.head()
```

```python
#exploring the data types and missing values
df.info()
```

Describtive Analysis

```python
df.describe()
```

```python
#fill the mising values in profits and votes by 0
df.budget_musd.fillna(0,inplace=True)
df.vote_count.fillna(0,inplace=True)
```

```python
df.hist(figsize=(20,12), bins =100)
```

```python
#majority of the rating score is between 5 and 8 
df['vote_average'].hist(figsize=(10,6),bins=50)
```

```python
#majority of movies have low budget
df['budget_musd'].hist(figsize=(10,6),bins=50, range=[0, 150])
plt.xlabel('budget in millions')
```

```python
#relationship between budget amount and rating
sns.regplot(df['budget_musd'],df['vote_average'],scatter_kws={"color": "blue"}, line_kws={"color": "red"})
```

```python
#relationship between revnue amount and rating
sns.regplot(df['revenue_musd'],df['vote_average'],scatter_kws={"color": "blue"}, line_kws={"color": "red"})
```

```python
#relationship between run time amount and rating
sns.regplot(df['runtime'],df['vote_average'],scatter_kws={"color": "blue"}, line_kws={"color": "red"})
```

```python
#relationship between run time amount and revenue
sns.regplot(df['runtime'],df['revenue_musd'],scatter_kws={"color": "blue"}, line_kws={"color": "red"})
```

```python
# adding profit column 
df['profit'] = df['revenue_musd'] - df['budget_musd']
```

```python
# top profit generating movies
top5 = df[['title','profit']].sort_values('profit', ascending=False).head()
plt.figure(figsize=(10,6))
sns.barplot(x=top5['title'],y=top5['profit'])
```

```python
from IPython.display import HTML
```

```python
top5p = df[['title','profit','poster_path']].sort_values('profit', ascending=False).head()
HTML(top5p.to_html(escape=False))
```

```python
#top 5 unprofitable genres by count of movies
unprofitable = df[df['profit']<0][['title', 'genres']].groupby('genres').count().reset_index()
top5_unprofitable = unprofitable.sort_values('title', ascending=False).head()
plt.figure(figsize=(10,6))
sns.barplot(x=top5_unprofitable['genres'],y=top5_unprofitable['title'])
```

```python
# movies with duplicate name 
df[df['title'].duplicated()==1]['title'].count()
```

```python
#count of each genre
df2 = df.copy()
gen = ['Animation','Comedy','Family','Adventure','Fantasy','Romance','Comedy','Crime','Horror']
for i in gen:
    df2[i] = df.genres.str.contains(i)*1
df2 = df2[['Animation','Comedy','Family','Adventure','Fantasy','Romance','Comedy','Crime','Horror']]
df2 = df2.melt(var_name="Gen", value_name="Count")
df2 = df2.groupby('Gen').sum().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=df2['Gen'],y=df2['Count'])
```

```python
# most common used words in the moview titles
from wordcloud import WordCloud
title = df.title.dropna()
one_str = ' '.join(title)
title_wordcloud = WordCloud(background_color='white', height=2000,
                            width = 4000, max_words=100).generate(one_str)
plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud, interpolation="bilinear")
plt.axis('off')
```

```python
# being part of series definitely affects the movie profit as an average
df['part_of_series'] = np.where(df['belongs_to_collection'].notna(),'series','single')
df[['profit','part_of_series']].groupby('part_of_series').mean().reset_index()
```

```python
# however, the voting doesn't look to have a big difference between single or series movies
df.groupby('part_of_series').agg(
    {'title':'count', 'budget_musd': ['sum','mean'],
     'revenue_musd':['sum','mean'], 'vote_average':'mean'})
```

```python
# most active directors
df['director'].value_counts().head(10)
```

```python
# most generating profits directors
direc = df[['profit','director']].groupby('director').sum().sort_values('profit',ascending=False).head(10).reset_index()
plt.figure(figsize=(20,6))
sns.barplot(x=direc['director'],y=direc['profit'])
```

```python
# move id column to be the index 
df.set_index('id', inplace=True)
```

```python
#seperate the actors of each movie to many columns 
act = df.cast.str.split('|',expand=True)
```

```python
act = act.stack().reset_index(level=1,drop=True).to_frame()
act.columns=['Actor']
```

```python
act = act.merge(df[['title', 'revenue_musd','vote_average','popularity']],
               how = 'left',left_index=True,right_index=True)
```
